```python
import btrdb
import pandas as pd
from matplotlib import pyplot as plt

db = btrdb.connect()
stream = db.stream_from_uuid('9cc032a1-6c32-4c3a-bb37-7c87da13dc4b')

# set query time
start, _ = stream.earliest()
dt = .1*1e9

# query database
points = stream.values(start.time, start.time+dt)

# transform data
times, values = zip(*[(pt.time, pt.value) for pt, _ in points])
data = pd.Series(values, index=times)
```

```python
# graph query results
data.plot()
```
